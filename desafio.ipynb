```python
import pandas as pd
import json
```

```python
df = pd.read_excel('planilha_vendas_produtos.xlsx')
df_limpo = df.drop(columns=['ID', 'CVM do Produto'], errors='ignore')
df_limpo = df_limpo.dropna()
```

```python
cliente_top = df_limpo.groupby('Nome Cliente')['Valor de Venda'].sum().idxmax()
valor_cliente = df_limpo.groupby('Nome Cliente')['Valor de Venda'].sum().max()

loja_top = df_limpo.groupby('Nome Loja')['Valor de Venda'].sum().idxmax()
faturamento_loja = df_limpo.groupby('Nome Loja')['Valor de Venda'].sum().max()

produto_top = df_limpo.groupby('Nome Produto')['Valor de Venda'].sum().idxmax()
receita_produto = df_limpo.groupby('Nome Produto')['Valor de Venda'].sum().max()

print(f"Melhor Cliente: {cliente_top} | Total Comprado: R$ {valor_cliente:,.2f}")
print(f"Melhor Loja: {loja_top} | Faturamento Total: R$ {faturamento_loja:,.2f}")
print(f"Produto Líder: {produto_top} | Receita Gerada: R$ {receita_produto:,.2f}")
```

```python
resultado_json = df_limpo.to_json(orient='records', force_ascii=False, indent=4)

with open('dados_vendas_ajustados.json', 'w', encoding='utf-8') as f:
    f.write(resultado_json)

print("\nETL Finalizado. Arquivo 'dados_vendas_ajustados.json' pronto para o banco.")
```
